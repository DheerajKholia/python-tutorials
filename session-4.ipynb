```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.value_counts()
```

```python
sns.displot(iris,x='sepal_length',binwidth=0.2)
```

```python
sns.displot(iris,x='sepal_length',hue='species')
```

```python
sns.displot(iris,x='sepal_length',hue='species',col='species')
```

```python
sns.displot(iris,x='sepal_length',hue='species',kind='kde')
```

```python
sns.catplot(data=iris,x='species',y='sepal_length',kind='box')
```

```python
sns.relplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```
