**Sample Problem** -
Write a program which will compute the minimun distance from origin to the circumference of a circle.
Length of the green line.

image.png

```python
## Given x, y coordinate of center. r = radius of circle.
x = 3
y= 4
r = 1
distance_center_from_origin = (x**2 + y**2) ** 0.5
print('distnace of center from origin : ', distance_center_from_origin)
minimun_distance = distance_center_from_origin - r
print('minimun distance : ', minimun_distance)
```

Python has strict formatting rule intended with spaces

```python
def calculate_distance(x, y, r):
  distance = (x**2 + y**2)**0.5 -r
  print(distance)
  return distance
print(calculate_distance(3, 4, 1))
```

```python
import math
math.hypot(3, 4) - 1

def cal_dist(x, y, r):
  result = math.hypot(x,y) -r
  return result
cal_dist(3, 4 ,1)
```
